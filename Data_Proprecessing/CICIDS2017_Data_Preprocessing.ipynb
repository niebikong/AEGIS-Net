```python
#df = df[df.line_race != 0]
df=df.dropna(axis=0, how='any')
df=df.replace(',,', np.nan, inplace=False)
```

```python
df = df[df['Flow Bytes/s'] != "Infinity"]
```

```python
df = df[df['Flow Packets/s'] != "Infinity"] 
df = df[df['Bwd Avg Bulk Rate'] != "Infinity"] 
df = df[df['Bwd Avg Packets/Bulk'] != "Infinity"] 
df = df[df['Bwd Avg Bytes/Bulk'] != "Infinity"] 
df = df[df['Fwd Avg Bulk Rate'] != "Infinity"] 
df = df[df['Fwd Avg Packets/Bulk'] != "Infinity"] 
df = df[df['Fwd Avg Bytes/Bulk'] != "Infinity"] 
df = df[df['CWE Flag Count'] != "Infinity"] 
df = df[df['Bwd URG Flags'] != "Infinity"] 
df = df[df['Bwd PSH Flags'] != "Infinity"] 
df = df[df['Fwd URG Flags'] != "Infinity"] 
```

```python
# df['Flow Bytes/s'].str.contains('Infinity').any()
df['Flow Bytes/s'] = df['Flow Bytes/s'].astype(str)  # 转换为字符串类型
df['Flow Bytes/s'].str.contains('Infinity').any()
```

```python
df["Bwd Avg Bulk Rate"].replace(",,", np.nan, inplace=True)
```

```python
df["Bwd Avg Bulk Rate"].replace(",,", np.nan, inplace=True)
df["Bwd Avg Packets/Bulk"].replace(",,", np.nan, inplace=True)
df["Bwd Avg Bytes/Bulk"].replace(",,", np.nan, inplace=True)
df["Fwd Avg Bulk Rate"].replace(",,", np.nan, inplace=True)
df["Fwd Avg Packets/Bulk"].replace(",,",np.nan, inplace=True)
df["Fwd Avg Bytes/Bulk"].replace(",,", np.nan, inplace=True)
df["CWE Flag Count"].replace(",,", np.nan, inplace=True)
df["Bwd URG Flags"].replace(",,", np.nan, inplace=True)
df["Bwd PSH Flags"].replace(",,", np.nan, inplace=True)
df["Fwd URG Flags"].replace(",,", np.nan, inplace=True)
```

```python
df.dropna()
df["Flow Bytes/s"]=df["Flow Bytes/s"].astype("float64")
df['Flow Packets/s']=df["Flow Packets/s"].astype("float64")
df['Bwd Avg Bulk Rate']=df["Bwd Avg Bulk Rate"].astype("float64")
df['Bwd Avg Packets/Bulk']=df["Bwd Avg Packets/Bulk"].astype("float64")
df['Bwd Avg Bytes/Bulk']=df["Bwd Avg Bytes/Bulk"].astype("float64")
df['Fwd Avg Bulk Rate']=df["Fwd Avg Bulk Rate"].astype("float64")
df['Fwd Avg Packets/Bulk']=df["Fwd Avg Packets/Bulk"].astype("float64")
df['Fwd Avg Bytes/Bulk']=df["Fwd Avg Bytes/Bulk"].astype("float64")
df['CWE Flag Count']=df["CWE Flag Count"].astype("float64")
df['Bwd URG Flags']=df["Bwd URG Flags"].astype("float64")
df['Bwd PSH Flags']=df["Bwd PSH Flags"].astype("float64")
df['Fwd URG Flags']=df["Fwd URG Flags"].astype("float64")
```
